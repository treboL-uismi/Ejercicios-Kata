## Representación de un diccionario

```python
D = {'age': 30, 'height': 175}
print(D)
```

## For Loop en los diccionarios

```python
D = {'age': 30, 'height': 175}

for elemento in D:
    print(f"{elemento}: {D[elemento]}")
```

## Añadir nuevos elementos en los diccionarios

```python
D = {'age': 30, 'height': 175}

P = {'sex': 'male'}

D.update(P)

print(D)
```

## Eliminar elementos en los diccionarios

```python
D = {'age': 30, 'height': 175, 'sex': 'male'}

D.pop('sex')

print(D)
```

## Dictionary Comprehension

```python
L = ['age', 'height']
A = {value: 30 for value in L if value == 'age'}
B = {value: 175 for value in L if value != 'age'}

A.update(B)

print(A)
```

## Dictionary Zip

```python
V = [(x+1)**2 for x in range(10)]

L = [x+1 for x in range(5)]

    
print(dict(zip(L, V)))
```

# Imaginemonos que queremos añadir una nueva value en nuestra key...
Mediante este algoritmo podremos llegar a hacerlo.

```python
def merge(*dicts):
    merged = {}

    for dict in dicts:
      for key, value in dict.items():
        if key not in merged:
          merged[key] = [value]
        else:
          merged[key].append(value)
    return merged

print(merge({'A': 2, 'B': 3}, {'E': 5, 'A': 2}))
```
